```python
import numpy as np
import pandas as pd
```

## Processing Metrics into CSV

```python
datasets = ['P53-MDM2', 'ecoli_data', 'yeast_cell']
impu = ['missforestMetrics', 'probsMetrics']
```

```python
columnsList = list(pd.read_csv(f"{datasets[0]}/metrics/{datasets[0]}_{impu[0]}.csv").columns)
```

```python
columnsList.append('impuType')
columnsList.append('dataset')
```

```python
dfMetrics = pd.DataFrame(columns=columnsList)
```

```python
dfMetrics
```

```python
for dataset in datasets:
    for imp in impu:
        dfRead = pd.read_csv(f"{dataset}/metrics/{dataset}_{imp}.csv")
        #print(dataset, imp)
        dfRead['impuType'] = imp
        dfRead['dataset'] = dataset
        dfMetrics = pd.concat([dfMetrics, dfRead])
```

```python
dfMetrics[dfMetrics['dataset']=='yeast_cell']
```

```python
dfMetrics['Dynamic Accuracy'] = dfMetrics['Dynamic Accuracy'].replace('Cannot be calculated', np.nan)
```

```python
dfMetrics
```

```python
dfMetrics.to_csv('summaryMetrics2.csv')
```

```python
df_melted = pd.melt(dfMetrics,
                    id_vars=['Method',	'Binarization', 'impuType', 'dataset'],
                    value_vars=['Dynamic Accuracy', 'Accuracy', 'Precision', 'Recall', 'F1-Score'],
                    var_name='metricName',
                    value_name='metricNumber')
```

```python
df_melted['metricNumber'] = df_melted['metricNumber'].astype(float)
```

```python
df_melted
```

```python
df_melted.to_csv('summaryMetrics.csv')
```

## Binarizations Processing

```python
methods = df_melted['Binarization'].unique()
bins = ['og_bin', 'bin_onlyMissForest', 'bin_Prob_and_MissForest']
dfPlot = pd.DataFrame(columns=['gene', 'binary', 'count', 'dataset', 'binarization', 'impu'])

dfPlot2 = pd.DataFrame(columns=["dataset", "binarization", "impu", "binary", "freq"])
```

```python
# iterate dataset
for d in datasets:

  # iterate thr methods
  for m in methods:

    # interate binarizations
    for b in bins:
      # read dataset
      df_bin = pd.read_csv(f'{d}/binarizations/{d}_{m}_{b}.csv')

      # replace ? as -1
      df_bin = df_bin.fillna(-1)
      # turn df to int
      df_bin = df_bin.astype(int)

      df_long = pd.melt(df_bin.reset_index(), id_vars=['index'], var_name='gene', value_name='binary')

      df_result = df_long.groupby(['gene', 'binary']).size().reset_index(name='count')
      
      df_long['binarization'] = m
      df_long['impu'] = b
      df_long['dataset'] = d
      
      df_count_groups = df_long.groupby(['dataset','binarization', 'impu', 'binary']).size().reset_index(name='freq')

      df_whole_dataset_count = df_count_groups[["dataset", "binarization", "impu", "binary", "freq"]]

      # Optional: Sort for better readability
      df_result = df_result.sort_values(by=['gene', 'binary']).reset_index(drop=True)

      df_result['dataset'] = d
      df_result['binarization'] = m
      df_result['impu'] = b

      dfPlot = pd.concat([dfPlot, df_result], axis=0)
      dfPlot.reset_index(drop=True, inplace=True)
      
      dfPlot2 = pd.concat([dfPlot2, df_whole_dataset_count], axis=0)
      dfPlot2.reset_index(drop=True, inplace=True)
      
```

```python
dfPlot['binary'] = dfPlot['binary'].replace(-1, '?')
dfPlot2['binary'] = dfPlot2['binary'].replace(-1, '?')
```

```python
dfPlot.to_csv("binCountperGene.csv")
```

```python
dfPlot2.to_csv("binCountWholeDataset.csv")
```
