```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("performance_100x100_approach2.csv")
df1 = pd.read_csv("performance_200x200_approach2.csv")
```

```python
df
```

```python
threads = [1, 2, 4, 8, 16, 32, 48]
```

```python
for i in df:
    #print(i)
    
    timeExecution = np.array(df[i].values)
    timeExecution1 = np.array(df1[i].values)
    
    timeExecution = timeExecution[0] / timeExecution
    timeExecution1 = timeExecution1[0] / timeExecution1
    
    #print(timeExecution)
    
    plt.plot(threads, timeExecution, label='Speedup Approach 100x100 Matrix', color='red', marker='o')
    plt.plot(threads, timeExecution1, label='Speedup Approach 200x200 Matrix', color='blue', marker='o')
    plt.plot(threads, threads, label='Ideal Speedup', color='black')

    plt.xlabel("Threads")
    plt.ylabel("Speedup")
    plt.grid(True)
    plt.title(f"Approach 2 Performance of {i.upper()}")
    
    plt.legend()

    plt.savefig(f'{i}_app2.png')
    plt.show()
```
