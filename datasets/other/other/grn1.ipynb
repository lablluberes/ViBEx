```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("yeast_cdc28.csv", header=0)
```

```python
data
```

```python
original_data = []

for index, row in data.iterrows():
    gene_name = row.iloc[0]
    values = row.iloc[1:].values
    original = []
    
    for v in values:
        
        #print(v)
        
        if not np.isnan(v):
            
            original.append(2 ** v)
        else:
           
            original.append(0)
        
    #break
 
    original_data.append([gene_name] + original)

# Convert list to DataFrame
df = pd.DataFrame(original_data)
```

```python
df
```

```python
df.to_csv("original_data_cdc28.txt", index=False, header=False, sep='\t')
```
