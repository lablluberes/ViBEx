Data was cleaned and converted to a single matrix and saved as CSV. Also, control samples were selected. 

```python
import pandas as pd
import numpy as np
```

```python
# import dataset
data = pd.read_csv('GSE8702_series_matrix.txt', sep="\t")
```

```python
data
```

```python
# read information txt file with gene names
labels = pd.read_csv('GPL570-55999.txt', sep="\t")
```

```python
labels
```

```python
# get a label dictionaty with id as key and gene name or symbol as key
labels_dict = labels.set_index('ID')['Gene Symbol'].to_dict()
```

```python
labels_dict
```

```python
data_new = data.copy()
```

```python
# create new column with gene name 
data_new["ID_REF"] = data["ID_REF"].replace(labels_dict)
```

```python
data_new
```

```python
# select control samples of the dataset
data2 = data_new[['ID_REF', 'GSM215632', 'GSM215633', 'GSM215636', 'GSM215639', 'GSM215642']]
```

```python
data2
```

```python
# save dataset with gene names
data2.to_csv('data_labels.csv')
```

```python
# only matrix of gene expressions
data = data2[['GSM215632', 'GSM215633', 'GSM215636', 'GSM215639', 'GSM215642']]
```

```python
data
```

```python
# save matrix of gene expression
data.to_csv('matrix.csv', index=False, header=False)
```
